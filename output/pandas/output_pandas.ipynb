```python
# dfのindex名が「3」のname列を
# '灼熱グラタン'に変更し、先頭の5行を表示

df.loc[3, 'name'] = '灼熱グラタン'
df.head()
```

```python
# dfのweek列にて月→0、火→1　水→２.....に
# 変更し、先頭の5行を表示

df['week'][df['week'] == '月'] = 0
df['week'][df['week'] == '火'] = 1
df['week'][df['week'] == '水'] = 2
df['week'][df['week'] == '木'] = 3
df['week'][df['week'] == '金'] = 4

df.head()
```

```python
# dfのtemperature列に100を足す

df['temperature'] = df['temperature'] +100

df['temperature']
```

```python
# dfのtemperature列に2を掛けて、更に気温をあげる
# 先頭の5行を表示

df['temperature'] = df['temperature'] * 2
df['temperature']
```

```python
# dfのtemperature列を小数点以下で丸める

df['temperature'] = df['temperature'].round()
df['temperature']
```

```python
# dfに列名「test」で値がすべて1の

df['test'] = 1
df.head()
```

```python
# dfにweatherとtemperatureの列を「_」で
# 結合した列を追加(列名は「test」)

df['temperature'] = df['temperature'].astype(str)

df['test'] = df['weather'] + ' : ' + df['temperature']
df.head()

```

```python
# dfにnameとsoldoutの列を「_」で
# 結合した列を追加(列名は「test2」)

df['test2'] = df['name'] + '--' + df['soldout'].astype(str)
df.head()
```

```python
# dfからtest, test2の列を削除し、最初の5行を表示

df.drop(['test', 'test2'], axis=1).head()
```

```python
# dfからインデックス名「3」の行を削除し、最初の5行を表示

df = df.drop(3, axis=0)
df.head()
```

```python
# 列名y を　Y(大文字)に変更

df = df.rename(columns={'y': 'Y'})
df.head()
```

```python
# dfのインデックス名「1」を「10」に変更

df = df.rename(index={1: 10})
df.head()
```

```python
# dfのすべての列の欠損値数を確認
df.isnull().sum()
```

```python
# dfのkcal列の欠損値に30を代入

df['kcal'] = df['kcal'].fillna(30)
df.head()
```

```python
# dfのsoldout列をndarray形式(配列)で表示

df['soldout'].values
```

```python
# dfの行をシャッフルして表示

df.sample(frac=1).head()
```

```python
# dfの行をシャッフルし、インデックスを振り直して表示
df.sample(frac=1).reset_index().head()
```

```python
# dfのweather列に含まれる「晴れ」という単語を
# 「灼熱」に置換。その後、1行目を確認

df['weather'] = df['weather'].replace('快晴', '灼熱')
df.head()
```

```python
# dfのname列1行目の「厚切りイカフライ」の[厚切り]を消去

import re
df['name'][0] = re.sub('厚切り', '', df['name'][0])
df['name'][0]
```

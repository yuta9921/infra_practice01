# 棒グラフの作成

```python
#簡単な棒グラフの作成
volume = [10,20,30,20,10]
labels = ['A', 'B', 'C', 'D', 'E']

#plt.bar(ｘ軸、ｙ軸)
plt.bar(labels, volume)
plt.show()
```

```python
#太さの変更
#0から１の間で宣言
plt.bar(labels, volume, width=0.5)
plt.show()
```

```python
#水平棒グラフ
plt.barh(labels, volume)
plt.show()
```

```python
#色の変更
plt.bar(labels, volume, width=0.5, color='red')
plt.show()
```

```python
#枠線の追加
plt.bar(labels, volume, width=0.5, edgecolor='red')
plt.show()
```

```python
#枠線の太さ
plt.bar(labels, volume, width=0.5, edgecolor='red', linewidth=5)
plt.show()
```

```python
#軸ラベル,タイトルの追加
plt.bar(labels, volume, width=0.5, edgecolor='red', linewidth=5)
plt.xlabel('Category')
plt.ylabel('Volume')
plt.title('BarGraph')
plt.show()
```

```python
#グラフ全体のサイズ
plt.figure(figsize=(10,6))
plt.bar(labels, volume, width=0.5, edgecolor='red', linewidth=5)
plt.xlabel('Category')
plt.ylabel('Volume')
plt.title('BarGraph')
plt.show()
```

```python
#グラフの保存
plt.figure(figsize=(10,6))
plt.bar(labels, volume, width=0.5, edgecolor='red', linewidth=5)
plt.xlabel('Category')
plt.ylabel('Volume')
plt.title('BarGraph')
plt.savefig('BarGraph.png')
```

```python
#2つ以上のグラフ表示

labels = ['A', 'B', 'C', 'D', 'E']
man = [35,45,25,10,20]
woman = [20,25,35,40,30]

plt.bar(labels, man, width=0.4)
plt.bar(labels, woman, width=0.4)
plt.show()
```

```python
#位置をずらす

labels = ['A', 'B', 'C', 'D', 'E']
man = [35,45,25,10,20]
woman = [20,25,35,40,30]

#align で位置をずらすことが出来る align='edge' で右寄り
plt.bar(labels, man, width=0.4, align='edge')
plt.bar(labels, woman, width=0.4, align='center')
plt.show()
```

```python
#凡例をつける

plt.bar(labels, man, width=0.4, align='edge', label='Man')
plt.bar(labels, woman, width=0.4, align='center', label='Woman')
plt.legend()
plt.show()
```

```python
#ラベルとタイトルをつける

plt.bar(labels, man, width=0.4, align='edge', label='Man')
plt.bar(labels, woman, width=0.4, align='center', label='Woman')
plt.legend()
plt.xlabel('Category')
plt.ylabel('Number of peaple ')
plt.title('Man and Woman')

plt.show()
```

```python
#積み上げ棒グラフの作成

plt.bar(labels, man, width=0.4, label='Man')
#woman のbottom（底）をman とすることで積み上げて表示できる　alignは消去
plt.bar(labels, woman, width=0.4, label='Woman', bottom=man)
plt.legend()
plt.xlabel('Category')
plt.ylabel('Number of peaple ')
plt.title('Man and Woman')

plt.show()
```

```python
#水平棒グラフの作成

plt.barh(labels, man,label='Man')
#bar を barh　に変更し、 left = manを宣言
plt.barh(labels, woman,label='Woman', left = man)

plt.legend()
plt.xlabel('Category')
plt.ylabel('Number of peaple ')
plt.title('Man and Woman')

plt.show()
```

```python
#水平棒グラフの太さ

#水平グラフの際はheightで太さを変更
plt.barh(labels, man, height=0.4, label='Man')
plt.barh(labels, woman, height=0.4, label='Woman', left = man)

plt.legend()
plt.xlabel('Category')
plt.ylabel('Number of peaple ')
plt.title('Man and Woman')

plt.show()
```

# 円グラフの作成

```python
#簡単な円グラフの作成
#合計が100じゃなくてもグラフを作ってくれる
value = [10,15, 25, 35]
label = ['tomato', 'carrot', 'potato', 'onion']

#plt.pie(x= 数値, labels= 各要素の名前)
plt.pie(x=value, labels=label)
plt.show()
```

```python
#％の表示

#小数第1位まで表示の宣言　'%1.1f%%'
plt.pie(x=value, labels=label, autopct='%1.1f%%')
plt.show()
```

 ## 円グラフ作成時にはデフォルトで3時の位置(0度)から反時計回りに値がふられる

```python
#12時の位置(90度)からスタートさせる

plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
#12時の位置(90度)から時計回りにスタートさせる

plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
#カラーの変更

#要素の数、色を宣言
colorlist=['pink', 'yellow', 'skyblue', 'orange']

plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False,
       colors=colorlist)
plt.show()
```

```python
#一部を切り出す

plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False,colors=colorlist, 
        explode=[0.1,0,0.3,0])#explodeに要素分の数字（０から１まで、今回は4つ）を与え数字が大きいほど飛び出す
plt.show()
```

```python
#立体感を出す

plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False,colors=colorlist, 
        explode=[0,0,0,0.3], shadow=True)
plt.show()
```

```python
#グラフにタイトルをつける

plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False,colors=colorlist, 
        explode=[0,0,0,0.3], shadow=True)
plt.title('PieGraph')
plt.show()
```

```python
#グラフを保存

plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False,colors=colorlist, 
        explode=[0,0,0,0.3], shadow=True)
plt.title('PieGraph')
plt.savefig('piegraph.png')
```

```python
#ドーナツ型のグラフ
#2つのグラフを作り重ねることでドーナツ型になる

plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False,colors=colorlist, 
        shadow=True)
#100%のグラフに背景と同じ白色を与え、radiusで大きさを指定(0～1を与える)。0.5で半分の大きさになる
plt.pie([100], colors='white', radius=0.5)

plt.title('PieGraph')
plt.show()
```

```python
#　%の位置をずらす
# pctdistance（０から１で指定）で％の表示場所を変更　値が大きいほど外側になる
plt.pie(x=value, labels=label, autopct='%1.1f%%', startangle=90, counterclock=False,colors=colorlist, 
        shadow=True, pctdistance=0.75)

plt.pie([100], colors='white', radius=0.5)

plt.title('PieGraph')
plt.show()
```

```python
#2つのグラフを横に並べて表示させる

x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

# plt.subplot(x y z) xが何行、ｙが何列、ｚが何番目
#今回は plt.subplot（１２１）なので1行２列の1番目にplt.plot(x, y_sin)を表示するという意味
plt.subplot(121)
plt.plot(x, y_sin)

#subplot（１２２）なので1行２列の２番目にplt.plot(x, y_cos)を表示するという意味
plt.subplot(122)
plt.plot(x, y_cos)

plt.show()
```

```python
#グラフ同士の余白を編集

x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(121)
plt.plot(x, y_sin)

plt.subplot(122)
plt.plot(x, y_cos)

# wspace(width-spaceの意味)０から１の数字を指定し大きいほどグラフ間の余白が広がる
plt.subplots_adjust(wspace=0.5)
plt.show()
```

```python
#全体のサイズを広げる

x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

#全体を横９　縦３の比に指定
plt.figure(figsize=(9,3))

plt.subplot(121)
plt.plot(x, y_sin)

plt.subplot(122)
plt.plot(x, y_cos)

plt.show()
```

### グラフの表示比率を変更

```python
from matplotlib import gridspec
```

```python
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

fig = plt.figure(figsize=(9,4))
# nrows は行、ncols は列を表す　 height_ratios=[2,1]は高さの比率を表す
#2行1列のグラフが生成され、1つ目と2つ目のグラフは2：１の大きさの割合で表示される
grid = gridspec.GridSpec(nrows=2, ncols=1, height_ratios=[2,1])

# grid[0] で1つ目（今回だと大きい方）に指定
ax1 = fig.add_subplot(grid[0])
ax1.plot(x, y_sin)

ax2 = fig.add_subplot(grid[1])
ax2.plot(x, y_cos)

plt.show()
```

```python
#グラフを横に並べる
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

fig = plt.figure(figsize=(9,4))
# nrows は行、ncols は列を表す　 height_ratios=[2,1]は高さの比率を表す
#1行2列のグラフが生成され、1つ目と2つ目のグラフは2：１の大きさの割合で表示される
grid = gridspec.GridSpec(nrows=1, ncols=2, width_ratios=[2,1])

# grid[0] で1つ目（今回だと大きい方）に指定
ax1 = fig.add_subplot(grid[0])
ax1.plot(x, y_sin)

ax2 = fig.add_subplot(grid[1])
ax2.plot(x, y_cos)

plt.show()
```
